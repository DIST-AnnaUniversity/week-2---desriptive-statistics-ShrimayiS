```python
import pandas as pd
```

```python
df = pd.read_csv("./wpi_monthly_data.csv")
```

```python
df.head()
```

```python
cols = df.columns.tolist()
```

```python
cols.remove('COMM_NAME')
cols.remove('COMM_CODE')
cols.remove('COMM_WT')
```

```python
for c in cols:
    try:
        print(f"Mean of {c} is {df[c].mean()}")
        print(f"Median of {c} is {df[c].median()}")
        print(f"Mode of {c} : ")
        print(df[c].mode())
        print("--" * 50)
    except Exception as e:
        pass
```
